```python
features = [
    "stars", "folks", "watching", "downloads", "downstream_dependents", 
    "contributors", "commit", "issue", "pull_request", "version", 
    "programming_language", "dependencies", "doc", "code"
]

dimensions = [
    "number", "time", "open/close", "format", "size", 
    "utilization", "comment", "security"
]

def enumerate_feature_combinations(features, dimensions):
    combinations = []
    for feature in features:
        for i in range(len(dimensions)):
            for j in range(i + 1, len(dimensions)):
                combination = f"{feature} ({dimensions[i]}, {dimensions[j]})"
                combinations.append(combination)
    return combinations

potential_metric_combinations = enumerate_feature_combinations(features, dimensions)
potential_metric_combinations
```

```python
# Raw metric categories

# We start with basic repo features first.
basic_metrics = ["stars", "forks", "subscribers", "downloads", "dependents", "contributors", "commits", "issues", 
 "pulls", "PL", "dependencies", "version", "code", "doc", "actions", "comment", "security"]
# comment indicates the comments of issue and pulls


# These features are mostly about cosmetic information, e.g., whether the commit messages are well formatted and detailed, whether the code has enough comments. 
cosmetic_metrics = ["commits", "issues", "pulls", "docs", "code"] # code here mean in-code comments.

# this metrics involve code.
code_metrics = ["commits", "pulls"]

# this metrics involve timestamps. 
longitudinal_metrics = ["commits", "issues", "pulls", "actions", "version", "comments"]

# this metrics involve contributors.
contributor_metrics = ["stars", "folks", "subscribers", "commits", "issues", "pulls", "downloads"]

# special metrics. we will need to enumerate more. we possiblely need to put all the external metrics here.
special_metrics = ["PL", "dependencies", "dependents"]
```

```python
# define enumeration metrics. The first letter indicate the data types. b means binary, n means numeric.
cosmetic_combos = ["b_msg_format", "n_msg_length", "n_avg_tags", "n_headlines", "n_coverage"] # n_coverage refers to, i.e., number of existing doc verses total docs in github requirement

code_combos = ["n_total_code_length", "n_avg_code_length", "n_avg_num_of_involved_files"]

longitudinal_combos = ["n_earliest_time", "n_avg_time","n_most_recent_time", "n_totalnum", "n_percentage_succeed", "n_frequency"] # n_percentage_succeed refers to whether a issue is closed, whether a pull request is merged, etc. 

# note we define contributors broadly here. A participant that involves in an issue discussion is also a contributor.
contributor_combos = ["n_total_contributor", "n_avg_contributor", "n_unique_contributor", "n_recurring_contributors"]  

# Unique Contributor Count metric refers to the total number of distinct individuals 
# who have contributed to a project over a specific period

special_combos = ["n_percentage_consist", "n_totalnum"]
```

```python
all_metrics = []
for metric1 in cosmetic_metrics: 
    for metric2 in cosmetic_combos:
        all_metrics.append(metric1 + "_" + metric2)
        
        
for metric1 in code_metrics: 
    for metric2 in code_combos:
        all_metrics.append(metric1 + "_" + metric2)
        
        
for metric1 in longitudinal_metrics: 
    for metric2 in longitudinal_combos:
        all_metrics.append(metric1 + "_" + metric2)        
        

for metric1 in contributor_metrics: 
    for metric2 in contributor_combos:
        all_metrics.append(metric1 + "_" + metric2) 

for metric1 in special_metrics: 
    for metric2 in special_combos:
        all_metrics.append(metric1 + "_" + metric2) 
        

# all_metrics.append() # TODO: Add special metrics here. 
print(all_metrics)
print(len(all_metrics))
```
